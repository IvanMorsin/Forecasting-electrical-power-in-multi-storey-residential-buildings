```python
import glob
import pandas as pd

# собираем результаты TFT
tft_files = sorted(glob.glob("results_tft_*.csv"))

df_tft = pd.concat([pd.read_csv(f) for f in tft_files], ignore_index=True)

for metric in ["MAE", "RMSE", "MAPE"]:
    pivot = df_tft.pivot(index="горизонт", columns="дом", values=metric)
    pivot = pivot.reindex(["4h", "8h", "24h", "7d", "14d", "1m"])
    print(f"\nTFT - {metric}:")
    print(pivot.to_string())
```

Результаты TFT подтверждают что модель хорошо работает только на коротких горизонтах - на 7d и 14d MAPE резко вырастает до 24-29%.

```python
df_tft.to_csv('results_tft.csv', index=False)
```
